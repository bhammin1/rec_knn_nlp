```python
import pandas as pd
import numpy as np
from scipy import spatial
from scipy import stats
import os
import os.path
import csv
import news_rec as nr
```

## Load all of the news and user data

```python
hist, can = nr.all_preprocessing()
```

```python
# filter experiments to specific users
exp1_3_users = can["user_id"].unique()
exp_users = exp1_3_users[0:1001]
user_file = './data/users.csv'
if os.path.isfile(user_file):
        print("File already exists")
        users = pd.read_csv(user_file)
else:
      users = pd.DataFrame(data={"user_id":exp_users})
      users.to_csv(user_file, sep=',', index=False)
users.head()
```

## Experiments

```python
sample_users = users["user_id"].to_list()
hist_sample = hist.loc[hist["user_id"].isin(sample_users)] # replaced users
can_sample = can.loc[can["user_id"].isin(sample_users)] # replaced users
```
